```python
player_input = widgets.Text(
    value='',
    placeholder='Player name',
    description='Player:',
    disabled=False
)
pos_input = widgets.Text(
    value='',
    placeholder='Position',
    description='Pos:',
    disabled=False
)
age_input = widgets.IntText(
    value=0,
    description='Age:',
    disabled=False
)
tm_input = widgets.Text(
    value='',
    placeholder='Team',
    description='Tm:',
    disabled=False
)
g_input = widgets.IntText(
    value=0,
    description='G:',
    disabled=False
)
gs_input = widgets.IntText(
    value=0,
    description='GS:',
    disabled=False
)
mp_input = widgets.FloatText(
    value=0.0,
    description='MP:',
    disabled=False
)
fg_input = widgets.FloatText(
    value=0.0,
    description='FG:',
    disabled=False
)
fga_input = widgets.FloatText(
    value=0.0,
    description='FGA:',
    disabled=False
)
fgp_input = widgets.FloatText(
    value=0.0,
    description='FG%:',
    disabled=False
)
tp_input = widgets.FloatText(
    value=0.0,
    description='3P:',
    disabled=False
)
tpa_input = widgets.FloatText(
    value=0.0,
    description='3PA:',
    disabled=False
)
tpp_input = widgets.FloatText(
    value=0.0,
    description='3P%:',
    disabled=False
)
tp2_input = widgets.FloatText(
    value=0.0,
    description='2P:',
    disabled=False
)
tp2a_input = widgets.FloatText(
    value=0.0,
    description='2PA:',
    disabled=False
)
tp2p_input = widgets.FloatText(
    value=0.0,
    description='2P%:',
    disabled=False
)
efgp_input = widgets.FloatText(
    value=0.0,
    description='eFG%:',
    disabled=False
)
ft_input = widgets.FloatText(
    value=0.0,
    description='FT:',
    disabled=False
)
fta_input = widgets.FloatText(
    value=0.0,
    description='FTA:',
    disabled=False
)
ftp_input = widgets.FloatText(
    value=0.0,
    description='FT%:',
    disabled=False
)
orb_input = widgets.FloatText(
    value=0.0,
    description='ORB:',
    disabled=False
)
drb_input = widgets.FloatText(
    value=0.0,
    description='DRB:',
    disabled=False
)
trb_input = widgets.FloatText(
    value=0.0,
    description='TRB:',
    disabled=False
)
ast_input = widgets.FloatText(
    value=0.0,
    description='AST:',
    disabled=False
)
stl_input = widgets.FloatText(
    value=0.0,
    description='STL:',
    disabled=False
)
blk_input = widgets.FloatText(
    value=0.0,
    description='BLK:',
    disabled=False
)
tov_input = widgets.FloatText(
    value=0.0,
    description='TOV:',
    disabled=False
)
pf_input = widgets.FloatText(
    value=0.0,
    description='PF:',
    disabled=False
)
pts_input = widgets.FloatText(
    value=0.0,
    description='PTS:',
    disabled=False
)

add_button = widgets.Button(
    description='Add Row',
    disabled=False,
    button_style='',
    tooltip='Click to add row',
    icon='plus'
)

# Function to handle button click
def add_row(b):
    global df
    new_row = {
        'Player': player_input.value,
        'Pos': pos_input.value,
        'Age': age_input.value,
        'Tm': tm_input.value,
        'G': g_input.value,
        'GS': gs_input.value,
        'MP': mp_input.value,
        'FG': fg_input.value,
        'FGA': fga_input.value,
        'FG%': fgp_input.value,
        '3P': tp_input.value,
        '3PA': tpa_input.value,
        '3P%': tpp_input.value,
        '2P': tp2_input.value,
        '2PA': tp2a_input.value,
        '2P%': tp2p_input.value,
        'eFG%': efgp_input.value,
        'FT': ft_input.value,
        'FTA': fta_input.value,
        'FT%': ftp_input.value,
        'ORB': orb_input.value,
        'DRB': drb_input.value,
        'TRB': trb_input.value,
        'AST': ast_input.value,
        'STL': stl_input.value,
        'BLK': blk_input.value,
        'TOV': tov_input.value,
        'PF': pf_input.value,
        'PTS': pts_input.value
    }
    df = df.append(new_row, ignore_index=True)
    print("Updated DataFrame:")
    display(df)

# Bind the button click event to the function
add_button.on_click(add_row)

# Display the widgets
display(player_input, pos_input, age_input, tm_input, g_input, gs_input, mp_input, fg_input, fga_input, fgp_input,
        tp_input, tpa_input, tpp_input, tp2_input, tp2a_input, tp2p_input, efgp_input, ft_input, fta_input, ftp_input,
        orb_input, drb_input, trb_input, ast_input, stl_input, blk_input, tov_input, pf_input, pts_input, add_button)
```

```python
playoffs_df = df[["Player","Pos","Age","Tm","AST","PTS","DRB","TRB","STL","BLK","G"]]
```

```python
playoffs_df
```

```python
playoffs_df['PTS_GEN'] = playoffs_df.apply(lambda row: row["AST"] + row["PTS"] + row["BLK"]+ (row["TRB"])*0.5+row["G"]*0.1, axis=1)
playoffs_df = playoffs_df.sort_values(by=['PTS_GEN'], ascending=False)
```
